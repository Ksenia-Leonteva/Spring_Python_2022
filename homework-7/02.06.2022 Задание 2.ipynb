### Exercise 1.1  

```python
x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x
fig, ax = plt.subplots()

ax.plot(x, y, color='green', label='y = x^2 * sin(1 / x^2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
```

### Exercise 1.2 

```python
x = np.linspace(5, 25, 250)
y = abs(1 - (1 / (1 + x**2)) / (1 / x**2))
fig, ax = plt.subplots()

ax.semilogy(x,y, color='green')
ax.set_title('Semilogy')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

### Exercise 1.3 

```python
x = np.linspace(-100, 100, 250)
y = scipy.special.dawsn(x)

fig, ax = plt.subplots()

ax.plot(x, y, color='green')
ax.set_title('Dawson`s integral')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
x = np.linspace(-50, 50, 250)
y = scipy.special.sinc(x)

fig, ax = plt.subplots()

ax.plot(x, y, color='green')
ax.set_title('The normalized sinc function')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6*np.pi, 6*np.pi, 250)
y = np.sin(x)/x
fig, ax = plt.subplots(figsize = (15, 4))

ax.plot(x, y, color='green', label='y = sin(x)/x')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid()

plt.xticks(np.linspace(-6*np.pi, 6*np.pi, 25),[r'$-6\pi$', r'$-11\pi/2$', r'$-5\pi$', r'$-9\pi/2$', r'$-4\pi$', r'$-7\pi/2$', 
                                               r'$-3\pi$', r'$-5\pi/2$', r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$',
                                               r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$', r'$5\pi/2$', r'$3\pi$', 
                                               r'$7\pi/2$', r'$4\pi$', r'$9\pi/2$', r'$5\pi$', r'$11\pi/2$', r'$6\pi$'])
ax.legend()
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

### Exercise 3.1 

```python
x = np.linspace(-np.pi, np.pi, 250)
y = np.linspace(-np.pi, np.pi, 250)
x, y = np.meshgrid(x,y)
z = np.sin(x) * np.sin(y)

fig, ax = plt.subplots(figsize=(7,7))
norm = mpl.colors.Normalize(-abs(z).max(), abs(z).max())
p = ax.pcolor(x, y, z, norm = norm,  cmap = mpl.cm.spring)

ax.axis('tight')
ax.set_xlabel(r"$x$", fontsize = 15)
ax.set_ylabel(r"$y$", fontsize = 15)
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
cb = fig.colorbar(p, ax = ax)
cb.set_label(r"$z$", fontsize = 15)
```

### Exercise 3.2 

```python
x = np.linspace(-3, 3, 250)
y = np.linspace(-3, 3, 250)
x, y = np.meshgrid(x,y)
z = np.exp(-x**2 - y**2)

fig, ax = plt.subplots(figsize = (7,7))
norm = mpl.colors.Normalize(-abs(z).max(), abs(z).max())
p = ax.pcolor(x, y, z, norm = norm,  cmap = mpl.cm.spring)

ax.axis('tight')
ax.set_xlabel(r"$x$", fontsize = 15)
ax.set_ylabel(r"$y$", fontsize = 15)
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(5))
cb = fig.colorbar(p, ax = ax)
cb.set_label(r"$z$", fontsize = 15)
```
