Бутылка Клейна в виде восьмёрки

```python
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.pyplot as plt
import numpy as np

u, v = np.linspace(0, 2*np.pi, 40), np.linspace(0, 2*np.pi, 40)
ux, vx =  np.meshgrid(u,v)
x = (1 + np.cos(ux/2) * np.sin(vx) - np.sin(ux/2) * np.sin(2*vx)) * np.cos(ux)
y = (1 + np.cos(ux/2) * np.sin(vx) - np.sin(ux/2) * np.sin(2*vx)) * np.sin(ux)
z = np.sin(ux/2) * np.sin(vx) + np.cos(ux/2) * np.sin(2*vx)

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111, projection = '3d')
plot = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = plt.get_cmap('viridis'))

for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)

plt.show()
```

```python
import plotly.graph_objects as go
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.pyplot as plt
import numpy as np


r = np.linspace(0.5, 1, 100)
p = np.linspace(0, 2*np.pi, 100)
rx, px = np.meshgrid(r,p)
x = -2 * rx * np.cos(px) + 2 * np.cos(px) / rx - 2 * rx**3 * np.cos(3*px) / 3
y = 6 * rx * np.sin(px) - 2 * np.sin(px) / rx - 2 * rx**3 * np.sin(3*px) / 3
z = 4*log(rx)

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111, projection = '3d')
plot = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = plt.get_cmap('rainbow'))

for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
    
plt.show()
```

```python
import plotly.graph_objects as go
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.pyplot as plt
import numpy as np

zz = np.linspace(-0.98, 0.98, 100)
tt = np.linspace(-np.pi, np.pi, 100)
zzx, ttx = np.meshgrid(zz, tt)

r = 4 * np.sqrt(1 - zzx**2) * (np.sin(abs(ttx)))**(abs(ttx))
x = r * np.sin(ttx)
y = r * np.cos(ttx)

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111, projection = '3d')
plot = ax.plot_surface(x, y, zzx, rstride = 1, cstride = 1, cmap = plt.get_cmap('spring'))

for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)
    
plt.show()
```

```python
import mpl_toolkits.mplot3d.axes3d as axes3d
import matplotlib.pyplot as plt
import numpy as np

u, v = np.linspace(-np.pi, np.pi, 400), np.linspace(-np.pi, np.pi, 400)
ux, vx =  np.meshgrid(u,v)
x = 2 * np.sin(3*ux) / (2 + np.cos(vx))
y = 2* (np.sin(ux) + 2 * np.sin(2*ux)) / (2 + np.cos(vx + 2 * np.pi / 3))
z = (np.cos(ux) - 2 * np.cos(2*ux)) * (2 + np.cos(vx)) * (2 + np.cos(vx + 2 * np.pi / 3)) / 4

fig = plt.figure()
fig.set_size_inches(10, 10)
ax = fig.add_subplot(111, projection = '3d')
plot = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = plt.get_cmap('viridis'))

for angle in range(0, 360):
    ax.view_init(30, angle)
    plt.draw()
    plt.pause(.001)

plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-3.14,3.14,1000)
x = np.cos(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12)**5)
y = np.sin(t) * (np.exp(np.cos(t)) - 2 * np.cos(4 * t) - np.sin(t / 12)**5)

fig, ax = plt.subplots()
ax.plot(x, y, color='green')
```
